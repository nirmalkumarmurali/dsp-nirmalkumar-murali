```python
import numpy as np

# Create a numpy array with shape (10, 3) filled with random integers between 0 and 100
array = np.random.randint(0, 100, size=(10, 3))

print("Generated array:")
print(array)
import matplotlib.pyplot as plt

# Plot the array
plt.figure(figsize=(8, 6))
plt.imshow(array, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Value')
plt.title('Random Integer Array Visualization')
plt.xlabel('Column')
plt.ylabel('Row')
plt.show()
```
